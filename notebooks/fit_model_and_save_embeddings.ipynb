# Initialize Word2Vec

# H E G E L

## English (Skip-Gram): remove_stop_words=False

```python
hegel_eng_path = Path('../', 'data', 'ENG_HEGEL_CORPUS.txt')
hegel_eng_raw = get_text(hegel_eng_path)

sg_vocab_freq_hegel_eng, sg_vocab_hegel_eng, sg_embeddings_hegel_eng = word2vec.fit(hegel_eng_raw, model_type='skipgram',
                                                                                    eng=True, de=False, remove_stop_words=False, 
                                                                                    sort_freq=15, window_size=3, n_epochs=3)
```

### Saving/Loading results in/from pkl

```python
with open('../embeddings/sg_hegel_eng_ws_3.pkl', 'wb') as f:  #wb - write binary 
    pickle.dump((sg_vocab_freq_hegel_eng, sg_vocab_hegel_eng, sg_embeddings_hegel_eng), f)

# with open('../embeddings/sg_hegel_eng.pkl', 'rb') as f:  #wb - write binary; rb - read binary
#     sg_vocab_freq_hegel_eng, sg_vocab_hegel_eng, sg_embeddings_hegel_eng = pickle.load(f)
```
